```python
from itertools import combinations
import pandas as pd

# Initialize nodes dictionary and edges list
nodes = {}
edges = []

# Group by sentence to identify interactions in the same sentence
grouped = final_df.groupby("sentence")[["Person", "House"]].apply(lambda x: x.values.tolist())

# Populate nodes and edges
for sentence, entities in grouped.items():
    # Add nodes with house information
    for person, house in entities:
        nodes[person] = house

    # Create edges for all pairs of entities in the same sentence
    persons_in_sentence = [person for person, _ in entities]
    edges.extend(combinations(persons_in_sentence, 2))

# Convert nodes into a DataFrame
nodes_df = pd.DataFrame(nodes.items(), columns=["Id", "House"])

# Prepare edges as a DataFrame, including a weight for the number of interactions
edges_df = pd.DataFrame(edges, columns=["source", "target"])
edges_df = edges_df.groupby(["source", "target"]).size().reset_index(name="Weight")

# Ensure only edges where both nodes exist in the nodes_df
valid_edges = edges_df[
    (edges_df["source"].isin(nodes_df["Id"])) &
    (edges_df["target"].isin(nodes_df["Id"]))
]

# Display the resulting DataFrames
print("Nodes DataFrame:")
print(nodes_df.head())

print("\nEdges DataFrame:")
print(valid_edges.head())
```
