```python
import pandas as pd
import spacy
from spacy import displacy
import requests
```

```python
# Load spaCy Transformer Model
nlp = spacy.load("en_core_web_trf")  # Transformer-based model
nlp.max_length = 2000000  # Handle larger texts

# URLs for the three books
urls = [
    "https://raw.githubusercontent.com/ganesh-k13/shell/master/test_search/www.glozman.com/TextPages/Frank%20Herbert%20-%20Dune.txt",
    "https://raw.githubusercontent.com/ganesh-k13/shell/master/test_search/www.glozman.com/TextPages/Frank%20Herbert%20-%20Dune%20Messiah.txt",
    "https://raw.githubusercontent.com/ganesh-k13/shell/master/test_search/www.glozman.com/TextPages/Frank%20Herbert%20-%20Children%20of%20Dune.txt"
]

# Fetch text from URLs
def fetch_text(url):
    response = requests.get(url)
    response.raise_for_status()  # Ensure the request was successful
    return response.text

# Fetch books and split them into lines
books = [fetch_text(url).splitlines() for url in urls]

# Remove specific rows for cleaning
books[0] = books[0][10:]  # Remove the first 10 rows from book 1
books[1] = books[1][5:]   # Remove the first 5 rows from book 2
books[2] = books[2][5:]   # Remove the first 5 rows from book 3

# Combine lines back into a single string for each book
books = ["\n".join(book) for book in books]

# Process each book individually
book_results = []

for book_number, book_text in enumerate(books, start=1):
    print(f"Processing Book {book_number}...")

    # Process the book text with spaCy
    doc = nlp(book_text)

    # Prepare a list to store sentence data
    book_data = []

    # Extract sentences and named entities
    for sent in doc.sents:
        entities = [{"text": ent.text, "label": ent.label_} for ent in sent.ents]
        book_data.append({
            "sentence": sent.text.strip(),
            "entities": entities
        })

    # Convert book data to a DataFrame and add the book column
    book_df = pd.DataFrame(book_data)
    book_df["book"] = f"book{book_number}"

    # Append the processed DataFrame to the results list
    book_results.append(book_df)

# Combine results from all books
final_df = pd.concat(book_results, ignore_index=True)

# Save the results to a CSV file
final_df.to_csv("books_transformer_ner_results.csv", index=False)

# Display a sample from the results
print(final_df.head())

# Visualize entities from a sample sentence in each book using spaCy's displacy
for book_number, book_text in enumerate(books, start=1):
    print(f"Visualizing NER for Book {book_number}...")
    doc = nlp(book_text[:1000])  # Use a smaller subset for visualization
    displacy.render(doc, style="ent", jupyter=True)
```

```python
# Extract sentences and person entities from processed text
person_entities_results = []

for book_number, book_text in enumerate(books, start=1):
    print(f"Processing Book {book_number} for PERSON entities...")
    doc = nlp(book_text)

    for sent in doc.sents:
        # Extract PERSON entities from the sentence
        person_entities = [ent.text for ent in sent.ents if ent.label_ == "PERSON"]
        # Append sentence and person entities to results
        person_entities_results.append({
            "book": f"Book {book_number}",
            "sentence": sent.text.strip(),
            "person_entities": ", ".join(person_entities) if person_entities else None
        })

# Convert the results into a DataFrame
person_entities_df = pd.DataFrame(person_entities_results)

# Save the DataFrame for further exploration
person_entities_df.to_csv("books_person_entities.csv", index=False)

# Display the first few rows of the DataFrame
print(person_entities_df.head())
```
